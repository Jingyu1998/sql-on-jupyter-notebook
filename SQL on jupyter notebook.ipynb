```python
# a = note b = query
def table_in_line(a,b):
    query = engine.connect().execute(b)
    df = pd.DataFrame([dict(i) for i in query])
    # 取出 join 後會變成浮點數的 column
    float_colname = df.select_dtypes(include = ['float64']).columns.tolist()
    # 將 浮點數 轉成 整數 
    # 將 空白值 補上 none
    df[float_colname] = df[float_colname].astype('Int32')
    df[float_colname] = df[float_colname].replace({np.nan: None})
    #並排顯示dataframe
    if a != "":
        # <br> 在 HTML檔用來換行
        df = df.style.set_table_attributes("style = 'display:inline'").set_caption(a + "<br>" + b)
    else:
        df = df.style.set_table_attributes("style = 'display:inline'").set_caption(b)
    return df

def table_in_line_sum(a,b):
    if len(a) != len(b):
        return None
    sum = table_in_line(a[0],b[0])._repr_html_()
    i = 1
    while i < len(a):
       sum = sum + table_in_line(a[i],b[i])._repr_html_()
       i = i + 1
    return sum
```

```python
##-- -----------------------------------------------
##-- select clause 檢視資料-------------------------
##-- ----------------------------------------------- 
##-- select * from customers
##-- 從 table customer 檢視 所有資料
##-- select last_name, first_name, points, (points + 10) * 100 as 'discount factor' from customers;
##-- 從 table customer 檢視column last_name, first_name, points, discount_factor的資料
##-- 其中 column discount factor 的資料是 (points 的資料 + 10) * 100 
note = ["", ""] 
query = ["select * from customers",
         "select last_name, first_name, points, (points + 10) * 100 as 'discount factor' from customers"]
#檢視表格
display_html(table_in_line_sum(note,query), raw=True)
```

```python
##-- -----------------------------------------------
##-- distinct 不顯示重複值 -------------------------
##-- -----------------------------------------------
##-- select distinct state from customers;-- 從 table customer 檢視 column state 的不重複資料 
note = ["", ""] 
query = ["select distinct state from customers",
         "select state from customers"]

display_html(table_in_line_sum(note,query), raw=True)
```

```python
##-- -----------------------------------------------
##-- like 搜尋字串 ---------------------------------
##-- -----------------------------------------------
##-- select * from customers where last_name like 'b%';
##-- 從 table customer 檢視 column last name 字首是b的所有資料
##-- ps % 代表0到無窮個字母
##-- select * from customers where last_name like '%b%';
##-- 從 table customer 檢視 column last name 字母有b的所有資料
##-- select * from customers where last_name like '%y';
##-- 從 table customer 檢視 column last name 字尾是y的所有資料
##-- select * from customers where last_name like 'b____y';
##-- 從 table customer 檢視 column last_name 總共六個字母，字首是b字尾是y的所有資料
##-- ps _ 代表 1 個字母

note = ["% 代表0到無窮個字母", "", "", "_ 代表 1 個字母"] 
query = ["select * from customers where last_name like 'b%'",
         "select * from customers where last_name like '%b%'",
        "select * from customers where last_name like '%y'",
        "select * from customers where last_name like 'b____y'"]

display_html(table_in_line_sum(note,query), raw=True)
```

```python
##-- -----------------------------------------------
##-- joining within database -----------------------
##-- 併同一資料庫的兩個table -----------------------
##-- -----------------------------------------------

##-- -----------------------------------------------
##-- inner join ------------------------------------
##-- 只顯示條件成立的資料---------------------------
##-- -----------------------------------------------
##-- select c.customer_id as c_c_id, o.customer_id as o_c_id , o.order_id from customers as c 
##-- join orders as o on c.customer_id = o.customer_id  order by c_c_id;
##-- 當 customers.customer_id 與 orders.customer_id 資料相同時，將 table orders 併在 table customers 的後方
##-- 只取部分 colummn

##-- -----------------------------------------------
##-- outer join (left join or right join )----------
##-- 將某一 table 的所有資料保留，條件不成立的資料以 null 補值
##-- -----------------------------------------------
##-- select c.customer_id as c_c_id, o.customer_id as o_c_id , o.order_id from customers as c 
##-- left join orders as o on c.customer_id = o.customer_id  order by c_c_id;
##-- 當 customers.customer_id 與 orders.customer_id 資料相同時，將 table orders 併在 table customers 的後方
##-- 不符合條件時以 null 補值，完整顯示執行 left join前 的資料
##-- 只取部分 colummn

note = ["customers table", "orders table", "inner join", "outer join"] 
query = ["select * from customers",
         "select * from orders as o order by o.customer_id",
         "select c.customer_id as c_c_id, o.customer_id as o_c_id , o.order_id from customers as c " + 
         "join orders as o on c.customer_id = o.customer_id  order by c_c_id",
         "select c.customer_id as c_c_id, o.customer_id as o_c_id , o.order_id from customers as c " + 
         "left join orders as o on c.customer_id = o.customer_id  order by c_c_id"]

display_html(table_in_line_sum(note,query), raw=True)
```

```python
##-- -----------------------------------------------
##-- join multiple tables 併多個table --------------
##-- -----------------------------------------------

##-- -----------------------------------------------
##-- inner join ------------------------------------
##-- -----------------------------------------------
##-- select c.customer_id as c_c_id, o.customer_id as o_c_id, o.shipper_id as o_s_id, s.shipper_id as s_s_id 
##-- from customers as c 
##-- join orders as o on c.customer_id = o.customer_id
##-- join shippers as s on o.shipper_id = s.shipper_id order by c_c_id;
##-- 當 customers.customer_id 與 orders.customer_id 資料相同時，將 table orders 併在 table customers 的後方
##-- 當 order.shipper_id 與 shipper.shipper_id 資料相同時，將 table shippers 併在 table orders 的後方
##-- 只選取部分column

##-- -----------------------------------------------
##-- outer join (left join or right join )----------
##-- -----------------------------------------------
##-- select c.customer_id as c_c_id, o.customer_id as o_c_id, o.shipper_id as o_s_id, s.shipper_id as s_s_id 
##-- from customers as c 
##-- left join orders as o on c.customer_id = o.customer_id
##-- left join shippers as s on o.shipper_id = s.shipper_id order by c_c_id;
##-- 當 customers.customer_id 與 orders.customer_id 資料相同時，將 table orders 併在 table customers 的後方
##-- 不符合條件時以 null 補值，完整顯示執行第一次 left join前 的資料
##-- 當 order.shipper_id 與 shipper.shipper_id 資料相同時，將 table shippers 併在 table orders 的後方
##-- 不符合條件時以 null 補值，完整顯示執行第二次 left join前 的資料
##-- 只選取部分column

note = ["customers table", "orders table","shippers table", "inner join", "outer join"] 
query = ["select * from customers",
         "select * from orders as o order by o.customer_id",
         "select * from shippers",
         "select c.customer_id as c_c_id, o.customer_id as o_c_id, o.shipper_id as o_s_id, s.shipper_id as s_s_id " + 
         "from customers as c " +
         "join orders as o on c.customer_id = o.customer_id " + 
         "join shippers as s on o.shipper_id = s.shipper_id order by c_c_id",
         "select c.customer_id as c_c_id, o.customer_id as o_c_id, o.shipper_id as o_s_id, s.shipper_id as s_s_id " + 
         "from customers as c " +
         "left join orders as o on c.customer_id = o.customer_id " + 
         "left join shippers as s on o.shipper_id = s.shipper_id order by c_c_id" ]
display_html(table_in_line_sum(note,query), raw=True)
```

```python
##-- -----------------------------------------------
##-- self join 併兩個相同的table -------------------
##-- -----------------------------------------------

##-- -----------------------------------------------
##-- inner join ------------------------------------
##-- 只顯示條件成立的資料--------------------------
##-- -----------------------------------------------
##-- select e.employee_id, e.first_name, m.first_name as manager from employees as e 
##-- join employees as m on e.reports_to = m.employee_id;
##-- 當 column e.reports_to 與 m.employee_id 相同時 
##-- 該筆資料複製到 table employee e 的後方
##-- 只選取部分column
##-- ------------------------------------------------

##-- -----------------------------------------------
##-- outer join (left join or right join )----------
##-- 將某一 table 的所有資料保留，條件不成立的資料以 null 補值
##-- -----------------------------------------------
##-- select e.employee_id, e.first_name, m.first_name as manager from employees as e 
##-- left join employees as m on e.reports_to = m.employee_id;
##-- 當 column e.reports_to 與 m.employee_id 相同時 
##-- 該筆資料複製到 table employee e 的後方
##-- 不符合條件時以 null 補值，完整顯示執行 left join前 的資料
##-- 只選取部分column
note = ["table employee","inner join", "outer join"] 
query = ["select employee_id, first_name, reports_to from sql_hr.employees",
         "select e.employee_id, e.first_name, m.first_name as manager from sql_hr.employees as e " +
         "join sql_hr.employees as m on e.reports_to = m.employee_id",
         "select e.employee_id, e.first_name, m.first_name as manager from sql_hr.employees as e " +
         "left join sql_hr.employees as m on e.reports_to = m.employee_id"]

display_html(table_in_line_sum(note,query), raw=True)
```

```python
##-- ------------------------------------------------
##-- using clause -----------------------------------
##-- 替代 on clause，更簡潔寫條件 -------------------
##-- 只能在 table們 的 column name 相同時使用 -------
##-- ------------------------------------------------
##-- select c.customer_id as c_c_id, o.customer_id as o_c_id , o.order_id from customers as c 
##-- join orders as o on c.customer_id = o.customer_id  order by c_c_id;
##-- select c.customer_id as c_c_id, o.customer_id as o_c_id , o.order_id from customers as c 
##-- join orders as o using (customer_id) order by c_c_id;
##-- 當 customers.customer_id 與 orders.customer_id 資料相同時，將 table orders 併在 table customers 的後方
##-- 只取部分 colummn


note = ["on clause", "using clause"] 
query = ["select c.customer_id as c_c_id, o.customer_id as o_c_id , o.order_id from customers as c " + 
         "join orders as o on c.customer_id = o.customer_id  order by c_c_id",
         "select c.customer_id as c_c_id, o.customer_id as o_c_id , o.order_id from customers as c " + 
         "join orders as o using (customer_id) order by c_c_id"]

display_html(table_in_line_sum(note,query), raw=True)
```
